```python
import os
import numpy as np
import pandas as pd
from shutil import copyfile
import serial
from gtts import gTTS
from io import BytesIO
import pygame
from keras.models import load_model
from tensorflow.keras.preprocessing import image
from skimage.transform import resize

model = load_model('D:\\Jupyter\\BrailleNet.h5')

from tensorflow.keras.preprocessing import image
from skimage.transform import resize

ser = serial.Serial('COM7', 9600)
letters = []
sentence = ""

# Create a dictionary to map data code to image file path
data_to_path = {
    '10000000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\a1.JPG0dim.jpg',
    '11000000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\b1.JPG0dim.jpg',
    '10010000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\c1.JPG0dim.jpg',
    '10011000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\d1.JPG0dim.jpg',
    '10001000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\e1.JPG0dim.jpg',
    '11010000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\f1.JPG0dim.jpg',
    '11011000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\g1.JPG0dim.jpg',
    '11001000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\h1.JPG0dim.jpg',
    '01010000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\i1.JPG0dim.jpg',
    '01011000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\j1.JPG0dim.jpg',
    '10100000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\k1.JPG0dim.jpg',
    '11100000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\l1.JPG0dim.jpg',
    '10110000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\m1.JPG0dim.jpg',
    '10111000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\n1.JPG0dim.jpg',
    '10101000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\o1.JPG0dim.jpg',
    '11110000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\p1.JPG0dim.jpg',
    '11111000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\q1.JPG0dim.jpg',
    '11101000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\r1.JPG0dim.jpg',
    '01110000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\s1.JPG0dim.jpg',
    '01111000': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\t1.JPG0dim.jpg',
    '10100100': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\u1.JPG0dim.jpg',
    '11100100': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\v1.JPG0dim.jpg',
    '01011100': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\w1.JPG0dim.jpg',
    '10110100': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\x1.JPG0dim.jpg',
    '10111100': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\y1.JPG0dim.jpg',
    '10101100': 'D:\\Personal\\College\\Official\\Project\\braille_map_img\\z1.JPG0dim.jpg'
}

while True:
    data = ser.readline().decode('utf-8').rstrip()
    print(data)
    
    if data in data_to_path:
        img_path = data_to_path[data]
        img = image.load_img(img_path)
        x = image.img_to_array(img)
        x = resize(x, (28, 28))
        x = np.expand_dims(x, axis=0)
        a = int(np.argmax(model.predict(x), axis=1))
        letter = chr(ord('`') + a + 1)
        print(letter)
        letters.append(letter)
        
    elif data == '00000010':
        # Space input
        letters.append(' ')
        
    elif data == '00000001':
        # Sentence input
        word = ''.join(letters)
        sentence += word + " "
        letters = []
        
    if len(sentence) > 0 and sentence[-1] == ' ':
        # Play the sentence audio when a stop is added
        mytext = sentence[:-1]
        print(mytext)
        sentence=""
        language = 'en'
        audio = BytesIO()
        gTTS(text=mytext, lang=language, slow=False).write_to_fp(audio)
        audio.seek(0)
        pygame.mixer.init()
        pygame.mixer.music.load(audio)
        pygame.mixer.music.play()
        while pygame.mixer.music.get_busy():
            continue
        pygame.mixer.quit()
```
