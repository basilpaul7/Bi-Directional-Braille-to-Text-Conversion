```python
import serial
import time

# Define the serial port and baud rate
serial_port = 'COM7'  # Update with the correct serial port
baud_rate = 9600

# Connect to the Arduino board
ser = serial.Serial(serial_port, baud_rate)
time.sleep(2)  # Wait for the Arduino to reset

# Define the Braille patterns for each letter (A-Z, a-z)
braille_patterns = {
    'A': '100000',
    'B': '110000',
    'C': '100100',
    'D': '100110',
    'E': '100010',
    'F': '110100',
    'G': '110110',
    'H': '110010',
    'I': '010100',
    'J': '010110',
    'K': '101000',
    'L': '111000',
    'M': '101100',
    'N': '101110',
    'O': '101010',
    'P': '111100',
    'Q': '111110',
    'R': '111010',
    'S': '011100',
    'T': '011110',
    'U': '101001',
    'V': '111001',
    'W': '010111',
    'X': '101101',
    'Y': '101111',
    'Z': '101011',
    'a': '100000',
    'b': '110000',
    'c': '100100',
    'd': '100110',
    'e': '100010',
    'f': '110100',
    'g': '110110',
    'h': '110010',
    'i': '010100',
    'j': '010110',
    'k': '101000',
    'l': '111000',
    'm': '101100',
    'n': '101110',
    'o': '101010',
    'p': '111100',
    'q': '111110',
    'r': '111010',
    's': '011100',
    't': '011110',
    'u': '101001',
    'v': '111001',
    'w': '010111',
    'x': '101101',
    'y': '101111',
    'z': '101011'
}

def send_braille_pattern(pattern):
    # Send the Braille pattern to the Arduino
    ser.write(pattern.encode())
    ser.write(b'\n')  # Send a newline character to signify the end of the pattern

# Main loop
while True:
    # Get user input
    user_input = input("Enter a letter (A-Z, a-z): ")
    
    if len(user_input) == 1 and user_input.isalpha():
        # Check if the letter has a defined Braille pattern
        if user_input in braille_patterns:
            # Send the Braille pattern to the Arduino
            send_braille_pattern(braille_patterns[user_input])
        else:
            print("Invalid letter. Please enter a letter (A-Z, a-z).")
    else:
        print("Invalid input. Please enter a single letter (A-Z, a-z).")
```
